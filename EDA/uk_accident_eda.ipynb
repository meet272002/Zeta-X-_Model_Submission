```python
df['date']=pd.to_datetime(df['date'],dayfirst=True,errors='coerce')
df['time']=pd.to_datetime(df['time'],errors='coerce',format='%H%M').dt.time
df['local_authority_district'].fillna(df['local_authority_highway_current'].mode()[0],inplace=True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

overall_data={
    'Total Accidents':len(df),
    'Total Casualties':df['number_of_casualties'].sum(),
    'Total Vehicles':df['number_of_vehicles'].sum(),
    'Fatal Accidents':(df['collision_severity']==1).sum(),
    'Serious Accidents':(df['collision_severity']==2).sum(),
    'Slight Accidents':(df['collision_severity']==3).sum()
}
plt.figure(figsize=(10,6))
plt.bar(overall_data.keys(),overall_data.values(),color='navy',alpha=0.7,edgecolor='black')
plt.title('Overall Accident Statistics')
plt.ylabel('Count')
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```

```python
sns.set_style("whitegrid")
plt.rcParams['figure.figsize']=(12,5)

severity_counts=df['collision_severity'].value_counts().sort_index()
severity_percent=df['collision_severity'].value_counts(normalize=True)*100

fig,ax = plt.subplots(1,2,figsize=(12,5))

ax[0].bar(['Fatal','Serious','Slight'],severity_counts.values,color=['lightcoral','orange','darkgreen'])
ax[0].set_title("Accident Severity Counts")
ax[0].set_xlabel('Severity')
ax[0].set_ylabel('COunt')

ax[1].pie(severity_percent.values,labels=['Fatal','Serious','Slight'],autopct='%1.1f%%',colors=['lightcoral','orange','darkgreen'],startangle=90)
ax[1].set_title("Severity Percentage Distribution")

plt.tight_layout()
plt.show()
```

```python
df['date']=pd.to_datetime(df['date'])
df['month']=df['date'].dt.month
monthly_counts=df.groupby('month').size().sort_index()

plt.figure(figsize=(12,6))
plt.plot(monthly_counts.index,monthly_counts.values,color='blue',marker='o',linewidth=2)
plt.fill_between(monthly_counts.index,monthly_counts.values,color='lightblue',alpha=0.3)
plt.title('Number of Accidents per Month')
plt.xlabel('Month')
plt.ylabel('Number of Accidents')
plt.xticks(range(1,13),['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'])
plt.grid(axis='y',linestyle='--',alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
df_Speed=df[df['speed_limit'].notnull()]
sns.boxplot(x='collision_severity',y='speed_limit',data=df_Speed,palette='Set2')
plt.title('Speed distribution by Accident Severity')
plt.xlabel('Severity(1=Fatal,2=Serious,3=Slight)')
plt.ylabel('Speed Limit (mph)')
plt.tight_layout()
plt.show()
```

```python
junction_map={13:'T or Staggered junction',16:'Crossroads',18:'Private drive or Entrance'}
filtered_df=df[df['junction_detail'].isin([13, 16, 18])]
filtered_df['junction_name']=filtered_df['junction_detail'].map(junction_map)
junction_counts=filtered_df['junction_name'].value_counts()

plt.figure(figsize=(10,6))
plt.bar(junction_counts.index,junction_counts.values,color='teal')
plt.title('Number of collisions by Junction Type')
plt.xlabel('Junction Type')
plt.ylabel('Number of Collisions')
plt.tight_layout()
plt.show()
```
